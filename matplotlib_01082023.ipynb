```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,4,5,6]
y=[2,4,5,2,1]
```

```python

plt.scatter(x,y)
plt.xlabel("Price")
plt.ylabel('Land')
plt.title("new_plot",fontfamily = "din",fontsize = 22,color="blue")
```

```python
plt.hist(x,bins=2)
```

```python
plt.bar(x,y)
plt.show()
```
