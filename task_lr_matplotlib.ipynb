### **bonus** : perform linear regression and plot graph for csv file

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn 
```

```python
# Data Loading
df = pd.read_csv("C:\\Users\\SHAIKH\\Desktop\\analyticsmile\\EDU\\Python\\dataset_files\\price.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
x = df[["Area (sqft)"]]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
```

```python
# plot 1
plt.figure(figsize=(8, 4),facecolor='grey')
plt.subplot(1,2,1)

ax = plt.axes()
ax.set_facecolor("black")
plt.scatter(x_test,y_test,color = 'r',marker="*")
plt.grid()
plt.xlabel('Area in sqft')
plt.ylabel("Price")
plt.title("Area vs Price Distribution")

# plot 2
plt.figure(figsize=(8, 4),facecolor='grey')
plt.subplot(1,2,2)

ax = plt.axes()
ax.set_facecolor("black")
plt.scatter(x_test,y_test,color = 'r',marker="h")
plt.plot(x_test,y_pred,linestyle = 'dotted',linewidth = 4,c = "green")
plt.grid()
plt.xlabel('Area in sqft',fontweight='bold')
plt.ylabel("Price",fontweight='bold')
plt.title("Area vs Price Distribution",fontweight='bold')

plt.show()
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```
